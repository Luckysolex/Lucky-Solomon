```python
import numpy as np
```

```python
# Creating a 1D array from a list
one_d_array = np.array([1, 2, 3, 4, 5])
print(one_d_array)
```

```python
# Creating a 2D array (matrix)
two_d_array = np.array([[1, 2, 3], [4, 5, 6]])
print(two_d_array)
```

# Arrays with specific values

```python
# Arrays with zeros
zeros = np.zeros((2, 3))
print(zeros)
```

```python
# Arrays with ones
ones = np.ones((3, 2))
print(ones)
```

```python
# Arrays with a specific value
value = np.full((2, 3), 7)
print(value)
```

# Broadcasting

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
print(a + 10)
```

```python
print(b * 10)
```

```python
# Broadcasting 1D to 2D
c = np.array([[1], [2], [3]])
d = np.array([4, 5, 6])
print(c)
print(d)
print(c + d)
```

# Statistical and Mathematical Functions

```python
data = np.array([1, 2, 3, 4, 5])

# Mean and Standard Deviation
print(np.mean(data))
print(np.std(data))
```

```python
# Sum along axis
matrix = np.array([[1, 2], [3, 4]])
print(np.sum(matrix, axis=0)) # column sums
print(np.sum(matrix, axis=1)) # row sums
```

```python
# Matrix inversion
inv_matrix = np.linalg.inv(matrix)
print(inv_matrix)
```

```python
# Eigenvalues and Eigenvectors
eigvals, eigvecs = np.linalg.eig(matrix)
print(eigvals)
```
